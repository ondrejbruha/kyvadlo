## Numerické řešení:




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

g = 9.81
l = 1.0
theta0 = np.pi / 4
omega0 = 0.0

def pendulum(t, y):
    theta, omega = y
    dtheta_dt = omega
    domega_dt = - (g / l) * np.sin(theta)
    return [dtheta_dt, domega_dt]

y0 = [theta0, omega0]

t_span = (0, 10)
t_eval = np.linspace(*t_span, 1000)

solution = solve_ivp(pendulum, t_span, y0, t_eval=t_eval, method='RK45')

theta = solution.y[0]
omega = solution.y[1]
time = solution.t

plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(time, theta)
plt.title('Graf kyvadla')
plt.xlabel('Čas')
plt.ylabel('Úhel')

plt.subplot(2, 1, 2)
plt.plot(time, omega)
plt.xlabel('Čas')
plt.ylabel('Úhlová rychlost')

plt.tight_layout()
plt.show()
```
